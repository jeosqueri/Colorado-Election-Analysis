```python
import pandas as pd
```

```python
results_csv = "2016Results.csv"

results_df = pd.read_csv(results_csv)

results_df.head(10)
```

## Clean Dem Presidential Results 2016

```python
dem_party = results_df.loc[results_df['PARTY'] == 'Democratic Party']
dem_party
```

```python
dem_presidential = dem_party.loc[dem_party['OFFICE / BALLOT ISSUE'] == 'President/Vice President']
dem_presidential
```

```python
total_dem_pres = dem_presidential.iloc[[64]]
total_dem_pres
```

```python
dem_counties = dem_presidential[0:64]
dem_counties
```

```python
dem_counties.to_csv('DemCountyResults2016.csv')
```

## Clean Rep Presidential Results 2016

```python
rep_party = results_df.loc[results_df['PARTY'] == 'Republican Party']
rep_party
```

```python
rep_presidential = rep_party.loc[rep_party['OFFICE / BALLOT ISSUE'] == 'President/Vice President']
rep_presidential
```

```python
total_rep_pres = rep_presidential.iloc[[64]]
total_rep_pres
```

```python
rep_counties = rep_presidential[0:64]
rep_counties
```

```python
rep_counties.to_csv('RepCountyResults2016.csv')
```
